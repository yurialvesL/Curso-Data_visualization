## Teste de hipótese

**H<sup>null</sup>**
**A distribuição da taxa da gorjeta é a mesma nos dois grupos**
**H<sup>alt</sup>**
**A distribuição da taxa da gorjeta não é a mesma nos dois grupos**

```python
from scipy.stats import ranksums
```

```python
sobremesa = gorjetas.query("sobremesa == 'Sim'").porcentagem
```

```python
sem_sobremesa = gorjetas.query("sobremesa == 'Não'").porcentagem
```

```python
"""Só pode ser aceito hipotese alternativa quando o pvalue < 0.5"""
r = ranksums(sobremesa,sem_sobremesa).pvalue 
```

```python
print(f'O valor do p-value é {round(r,2)}')
```

**H<sup>null</sup>**
<br>
	**A distribuição da taxa da gorjeta é a mesma nos dois grupos**

 ## Teste de hipótese

**H<sup>null</sup>**
**A distribuição do valor da conta é igual no sábado e no domingo**

**H<sup>alt</sup>**
**A distribuição do valor da conta não é igual no sábado e no domingo**

```python
valor_conta_domingo = gorjetas.query("dia_da_semana == 'Dom'").valor_conta
```

```python
r2 = round(ranksums(valor_conta_domingo,valor_conta_sabado).pvalue,2)
print(f'O valor do p-value é {r2}')
```

**H<sup>null</sup>**
**A distribuição do valor da conta é igual no sábado e no domingo**
